```python
import pandas as pd 
```

```python
# 불법복제물 오프라인 단속 현황 
data_illegal = pd.read_csv('data/(재)한국저작권보호원_불법복제물 오프라인 단속현황_20201021.csv', encoding='CP949')
data_illegal.sample(3)
```

```python
data_illegal = data_illegal.iloc[:, :10]
data_illegal = data_illegal.loc[data_illegal['불법복제물장르구분']=='음악', :]
data_illegal
```

```python
# 음악 불법 복제물 유통량 
illegal_distribute = pd.read_csv('data/한국저작권보호원_콘텐츠별 불법복제물 유통량(08년_17년).csv', encoding='CP949')
illegal_distribute = illegal_distribute.head(1)
illegal_distribute
```

```python
# 음악 온라인 불법 복제물 경험률 
illegal_exper = pd.read_csv('data/한국저작권보호원_콘텐츠별 온라인 불법복제물 경험률(08년_17년).csv', encoding='CP949')
illegal_exper = illegal_exper.head(1)
illegal_exper
```

```python
# 음악 불법 복제물 유통량과 복제물 경험률 concat 
illegal = pd.concat([illegal_distribute, illegal_exper])
illegal
```
